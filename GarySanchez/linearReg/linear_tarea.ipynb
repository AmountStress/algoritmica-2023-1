```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ARDRegression
from sklearn.linear_model import ElasticNetCV

from sklearn.model_selection import  train_test_split
```

```python
df = pd.read_csv('Clean_Dataset.csv')
df.describe()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.destination_city = df.destination_city.map({'Mumbai' : 0, 'Bangalore' : 1, 'Kolkata' : 2, 'Hyderabad' : 3, 'Chennai' : 4, 'Delhi' : 5})
```

```python
df.source_city = df.source_city.map({'Mumbai' : 0, 'Bangalore' : 1, 'Kolkata' : 2, 'Hyderabad' : 3, 'Chennai' : 4, 'Delhi' : 5})
```

```python
df.head()
```

```python
df.airline = df.airline.map({'SpiceJet' : 1, 'AirAsia' : 2, 'Vistara' : 3, 'GO_FIRST' : 4, 'Indigo' : 5,
       'Air_India' : 6})
df.airline.unique()
```

```python
df.stops = df.stops.map({'zero' : 0, 'one' : 1, 'two_or_more' : 2})
```

```python
df['class'] = df['class'].map({'Economy' : 0, 'Business' : 1})
clase = {0 : 'Economy', 1 : 'Business'}
```

```python
dfn = df.drop(['flight', 'departure_time', 'arrival_time'], axis=1)
corr = dfn.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
linear = LinearRegression()
ridge = Ridge()
ARDN = ARDRegression()
elastic = ElasticNetCV()
```

```python
data_x = df.drop(['price', 'flight', 'departure_time', 'arrival_time', 'duration'], axis=1)
data_y = df['price']
```

```python
data_x.head()
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
linear.fit(train_x, train_y)
elastic.fit(train_x, train_y)
ridge.fit(train_x, train_y)
ARDN.fit(train_x, train_y)
```

```python
print(f'y = {round( linear.coef_[0], 2)}x + {round(linear.intercept_, 2)}')
print(f'y = {round( ridge.coef_[0], 2)}x + {round(linear.intercept_, 2)}')
print(f'y = {round( elastic.coef_[0], 2)}x + {round(linear.intercept_, 2)}')
print(f'y = {round( ARDN.coef_[0], 2)}x + {round(linear.intercept_, 2)}')
```

```python
print(f'Score Linear reg: {round(linear.score(test_x, test_y), 10) * 100}%')
print(f'Score ARDN: {round(ARDN.score(test_x, test_y), 10) * 100}%')
print(f'Score Ridge reg: {round(ridge.score(test_x, test_y), 10) * 100}%')
print(f'Score ElasticCV: {round(ridge.score(test_x, test_y), 10) * 100}%')

```
